```python
#41.Write a Python program to check whether a list contains a sub list
def contains_sublist(main_list, sub_list):
    sub_len = len(sub_list)
    for i in range(len(main_list) - sub_len + 1):
        if main_list[i:i + sub_len] == sub_list:
            return True
    return False
main = [1, 2, 3, 4, 5]
sub = [3, 4]

result = contains_sublist(main, sub)
print("Sublist found?" , result)
```

```python
#61.Write a Python function to calculate the factorial of a number (a nonnegative integer)

def factorial(n):
    if n < 0:
        return "Factorial is not defined for negative numbers."

    result = 1
    for i in range(2, n + 1):
        result *= i
    return result

num = 5
print(f"Factorial of {num} is {factorial(num)}")
```

```python
#63.Write a Python function to check whether a number is perfect or not.

def is_perfect(number):
    if number <= 0:
        return False

    divisors_sum = 0
    for i in range(1, number):
        if number % i == 0:
            divisors_sum += i

    return divisors_sum == number

num = 28
if is_perfect(num):
    print(f"{num} is a perfect number.")
else:
    print(f"{num} is not a perfect number.")
```

```python
#78. Write a python program to find the longest words.

def find_longest_words(filename):
    with open(filename, 'r') as file:
        words = file.read().split()
        max_len = max(len(word) for word in words)
        longest_words = [word for word in words if len(word) == max_len]
    return longest_words

longest = find_longest_words("example.txt")
print("Longest word(s):")
for word in longest:
    print(word)
```

```python
#80.Write a Python program to count the frequency of words in a file.
from collections import Counter
import string

def count_word_frequency(filename):
    with open(filename, 'r') as file:
        text = file.read().lower()
        # Remove punctuation
        text = text.translate(str.maketrans('', '', string.punctuation))
        words = text.split()
        word_counts = Counter(words)
    return word_counts

frequencies = count_word_frequency("example.txt")
print("Word Frequencies:")
for word, count in frequencies.items():
    print(f"{word}: {count}")
```

```python
#19.Create a random 4x4 matrix and use NumPy to compute its eigenvalues and eigenvectors. Verify the eigenvalues by reconstructing the matrix.
import numpy as np

np.random.seed(0)
A = np.random.rand(4, 4)
print("Original Matrix A:\n", A)

eigenvalues, eigenvectors = np.linalg.eig(A)
print("\nEigenvalues:\n", eigenvalues)
print("\nEigenvectors:\n", eigenvectors)

D = np.diag(eigenvalues)
V = eigenvectors
V_inv = np.linalg.inv(V)
A_reconstructed = V @ D @ V_inv

print("\nReconstructed Matrix A from Eigen Decomposition:\n", A_reconstructed)

is_close = np.allclose(A, A_reconstructed)
print("\nReconstruction successful:", is_close)
```

```python
#21. Perform matrix multiplication of two 2D arrays using np.dot () and @. Compare the results and performance of both methods using a large dataset.

import numpy as np
import time

np.random.seed(42)
A = np.random.rand(1000, 1000)
B = np.random.rand(1000, 1000)

start_dot = time.time()
result_dot = np.dot(A, B)
end_dot = time.time()
time_dot = end_dot - start_dot
print("Time using np.dot():", time_dot, "seconds")

start_at = time.time()
result_at = A @ B
end_at = time.time()
time_at = end_at - start_at
print("Time using @ operator:", time_at, "seconds")

are_equal = np.allclose(result_dot, result_at)
print("Are results equal?:", are_equal)
```

```python
#22. Create a 3D array of shape (2, 1, 4) and a 2D array of shape (4, 1). Perform an element-wise operation using broadcasting and explain the result. Use np. new axis to achieve the same result without broadcasting.

A = np.array([[[1, 2, 3, 4]], [[5, 6, 7, 8]]])
print("3D Array A (shape {}):\n{}".format(A.shape, A))

B = np.array([[10], [20], [30], [40]])
print("\n2D Array B (shape {}):\n{}".format(B.shape, B))


result_broadcast = A + B.T
print("\nResult using broadcasting (A + B.T):\n", result_broadcast)

B_expanded = B.T[np.newaxis, :, :]
A_expanded = A
result_manual = A_expanded + B_expanded
print("\nResult using np.newaxis:\n", result_manual)

print("\nAre both results equal?:", np.array_equal(result_broadcast, result_manual))
```
